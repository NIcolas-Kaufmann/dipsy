# Plot the data-sets

```python
import dipsy
import numpy as np 
%matplotlib inline
```

```python
testi = dipsy.data.Testi2014()
f, ax = testi.plot()
ax.plot(0.023233972595735387*1000,3.805896680299848,"g+")
f.savefig('testi.pdf', transparent=True)
```

```python
tripathi = dipsy.data.Tripathi2017()
f, ax = tripathi.plot()
ax.plot( np.log10(0.024326021635537102),np.log10(13.298093822846921),"g+")
f.savefig('tripathi.pdf', transparent=True)
```
